```python
# Step 1: Import Required Libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Step 2: Load Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Step 3: Check for Missing Values
print("Missing Values:")
print(df.isnull().sum())

# Step 4: Describe the Data
print("\nStatistical Summary:")
print(df.describe())

# Step 5: Get Data Types and General Info
print("\nData Types and Info:")
print(df.info())

# Step 6: Check Dimensions of the DataFrame
print("\nShape of the DataFrame:")
print(df.shape)  # (rows, columns)

# Step 7: Check Current Data Types
print("\nColumn Data Types:")
print(df.dtypes)

# Step 8: Convert Data Type of 'pclass' (if needed)
df['pclass'] = df['pclass'].astype(str)

# Step 9: Convert Categorical Variables to Quantitative
# Convert 'sex' column: male = 0, female = 1
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# Convert 'embarked' column: S = 0, C = 1, Q = 2
df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Step 10: Check Updated Data Types and First Few Rows
print("\nUpdated Data Types After Conversion:")
print(df.dtypes)

print("\nFirst 5 Rows of the Updated Dataset:")
print(df.head())
```
