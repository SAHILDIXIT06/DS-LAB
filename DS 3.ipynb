```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target_names[data.target]  # Add species names

# Show summary statistics for 'sepal length (cm)' grouped by species
print(df.groupby('species')['sepal length (cm)'].agg(['mean', 'median', 'min', 'max', 'std']))

# Convert species to numeric values and show the list
df['species_num'] = pd.Categorical(df['species']).codes
print("\nNumeric values for species:\n", df['species_num'].tolist())

# Filter each species
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']

# Describe each
print("\nStatistics for Iris-setosa:")
print(setosa.describe())

print("\nStatistics for Iris-versicolor:")
print(versicolor.describe())

print("\nStatistics for Iris-virginica:")
print(virginica.describe())

```
